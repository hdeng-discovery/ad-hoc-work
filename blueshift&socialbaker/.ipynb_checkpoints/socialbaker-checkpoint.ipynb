```python
agg_metrics_call='/3/aggregated-metrics'
```

```python
def get_agg_metrics(platform,metric,start_date,end_date,label_id):
#     creds = get_api_creds()
    agg_metrics_call='/3/aggregated-metrics'
    call=json.loads(json.dumps(json.load(open(platform+'.json'))).\
                    replace('{1}',start_date).replace('{2}',end_date).replace('{0}',metric).replace('{3}',label_id))
    result=requests.post(base+agg_metrics_call, auth=(user, secret),json=call).json()
    dic={}
    df=pd.DataFrame()
    cnt=0
    print(result)
    for account in result['header'][0]['rows']:
        dates=0
        for day in result['header'][1]['rows']:
            dic['id']=account['id']
            dic['platform']=account['platform']
            dic['metric']=result['header'][2]['rows'][0]
            dic['date']=day
            dic['data']=result['data'][cnt][dates]
            df=df.append(pd.DataFrame.from_dict(dic))
            dates+=1
        cnt+=1
    return df
```

## followers

```python
def get_fans(platform,metric,date,profile_id):
    profile_call='/3/{0}/metrics'.format(platform)
    call=json.loads(json.dumps(json.load(open('fans.json'))).replace('{0}',date).replace('{1}',metric).replace('{2}',profile_id))
    result=requests.post(base+profile_call, auth=(user, secret),json=call).json()
    print(result)
    dic={}
    df=pd.DataFrame()
    cnt=0
    for account in result['header'][0]['rows']:
        dic['id']=profile_id
        dic['platform']=platform
        dic['metric']='Fans / Followers'
        dic['data']=result['data'][cnt][0]
        df=df.append(pd.Series(dic),ignore_index=True)
        cnt+=1
    df['timestamp']=datetime.datetime.now()
    return df
```

## post level data

```python
import re
def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"\u2019"
                           u"\U0001F600-\U0001F64F" 
                           u"\U0001F300-\U0001F5FF"
                           u"\U0001F680-\U0001F6FF"
                           u"\U0001F1E0-\U0001F1FF"
                           u"\U00002702-\U000027B0"
		                   u"\U000024C2-\U0001F251"
		                   u"\U0001f926-\U0001f937"
		                   u'\U00010000-\U0010ffff'
		                   u"\u200d"
		                   u"\u2640-\u2642"
		                   u"\u2600-\u2B55"
		                   u"\u23cf"
		                   u"\u23e9"
		                   u"\u231a"
		                   u"\u3030"
		                   u"\ufe0f"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r' ', string)

def get_post_metrics(platform,start_date,end_date,label_id='79fec95be74340998fea89bcac3c5de8'):
    call=json.loads(json.dumps(json.load(open(platform+'_post.json'))).\
                    replace('{1}',start_date).replace('{2}',end_date).replace('{0}',label_id))
    df=pd.DataFrame()
    if platform=='fb':
        path='Facebook/page/posts'
    elif platform=='ig':
        path='Instagram/profile/posts'
    elif platform=='yt':
        path='Youtube/profile/videos'
    result=requests.post(base+'/3/{0}'.format(path.lower()), auth=(user, secret),json=call).json()
    dic={}
    cnt=0
    print(result)
    for post in result['data']['posts']:
        dic['platform']=path.split('/')[0]
        dic['post_id']=post['id']
        dic['post_timestamp']=post['created_time']
        if platform=='yt':
            dic['content']=remove_emoji(str(post['description']))
        else:
            dic['content']=remove_emoji(str(post['content']))
        dic['post_url']=post['url']
        dic['author']=post['author']['name']
        dic['author_url']=post['author']['url']
        try:
            dic['insights_interactions']=post['interactions']
        except:
            dic['insights_interactions']='None'
        try:
            dic['insights_impressions']=post['insights_impressions']
        except:
            dic['insights_impressions']='None'
        if platform=='fb':
            dic['insights_post_clicks']=post['insights_post_clicks']
        else:
            dic['insights_post_clicks']='None'
        try:
            dic['insights_reach']=post['insights_reach']
        except:
            dic['insights_reach']='None'  
        if platform=='yt':
            try:
                dic['insights_video_views']=post['video_views']
            except:
                dic['insights_video_views']='None'
        else:
            try:
                dic['insights_video_views']=post['insights_video_views']
            except:
                dic['insights_video_views']='None'
        df=df.append(pd.Series(dic),ignore_index=True)
    df['timestamp']=datetime.now(timezone('EST')) 
    return df
```
